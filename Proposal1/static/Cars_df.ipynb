```python
import pandas as pd
import numpy as np
```

```python
# Store csv into df
csv_file= "../data/USA_cars_datasets_base.csv" 
cars_df1= pd.read_csv(csv_file, index_col=0)
cars_df1
```

```python
# column names
for col in cars_df1.columns:
    print(col)
```

```python
cars_df3= cars_df1
```

```python
indexNames = cars_df3[(cars_df3['price'] == 0)].index
cars_df3.drop(indexNames , inplace=True)
```

```python
indexNames1 = cars_df3[(cars_df3['country'] == ' canada')].index
cars_df3.drop(indexNames1 , inplace=True)
```

```python
cars_df= cars_df3
cars_df
```
